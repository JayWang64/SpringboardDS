# b) features that depend on cumulative sum

### Win Pct - Home Games

```python
# https://towardsdatascience.com/the-tale-of-missing-values-in-python-c96beb0e8a9d

# fillna wrt each season, team

df['c_WIN_pct'] = (df['c_WINS'] / df['c_GAMES'])

df['c_WIN_pct_at_HOME'] = (df['c_WINS_at_HOME'] /  df['c_GAMES_at_HOME'])
```

### Win Pct - Road aka Away Games

```python
df['c_WINS_at_AWAY'] = df['c_WINS'] - df['c_WINS_at_HOME']

df['c_GAMES_at_AWAY'] = df['c_GAMES'] - df['c_GAMES_at_HOME'] 

df['c_WIN_pct_at_AWAY'] = df['c_WINS_at_AWAY'] /  df['c_GAMES_at_AWAY']
```

### PTS differential per Game

```python
df['d_PTS_per_GAME'] = (df['c_PTS'] - df['c_PTS_opp']) / df['c_GAMES']
```

### FGM and FGA per Game

```python
df['c_FGM_per_GAME'] = (df['c_FGM']) / df['c_GAMES']
df['c_FGA_per_GAME'] = (df['c_FGA']) / df['c_GAMES']
```

### Pts For and Against per Game

```python
df['c_PTS_per_GAME'] = (df['c_PTS']) / df['c_GAMES']
df['c_PTS_opp_per_GAME'] = (df['c_PTS_opp']) / df['c_GAMES']
```

### Possessions
### https://www.nbastuffer.com/analytics101/possession/

```python
df['c_POS'] = 0.96 * ((df['c_FGA']) + (df['c_TO']) + (0.44 * df['c_FTA']) - (df['c_OREB']))
df['c_POS_opp'] = 0.96 * ((df['c_FGA_opp']) + (df['c_TO_opp']) + (0.44 * df['c_FTA_opp']) - (df['c_OREB_opp']))
```

### Offensive / Defensive Efficiency

### https://www.nbastuffer.com/analytics101/offensive-efficiency/
### https://www.nbastuffer.com/analytics101/defensive-efficiency/

aka Offensive Rating
https://www.sports-reference.com/cbb/about/glossary.html#:~:text=Offensive%20Rating%3B%20for%20players%20it,*%20(PTS%20%2F%20Poss).

```python
df['c_OFF_EFF'] = 100 * (df['c_PTS'] / df['c_POS'])
df['c_OFF_EFF_opp'] = 100 * (df['c_PTS_opp'] / df['c_POS_opp'])

df['c_DEF_EFF'] = 100 * (df['c_PTS_opp'] / df['c_POS'])
df['c_DEF_EFF_opp'] = 100 * (df['c_PTS'] / df['c_POS_opp'])
```
