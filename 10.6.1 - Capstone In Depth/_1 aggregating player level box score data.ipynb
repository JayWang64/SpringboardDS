# preparing the final group by using sum, grouped on GAME_ID, TEAM_ID
now we will have more interesting metrics such as: shots attempts/made, steals, blocks, fouls, etc. aggregated at the game level, per team.

```python
# here it is with the columns using the **index numbers**:

df.groupby(df.columns[0:2].values.tolist())[
    df.columns[9:].values.tolist()
].sum()
```

```python
# slicing the overall dataframe down to the attributes of interest
df_grp2 = df_grp[['GAME_ID', 'TEAM_ID', 'FGM', 'FGA','FG3M', 'FG3A', 'FTM','FTA', 'OREB','DREB','TO','PF']]
df_grp2
```

```python
# perform the merge -- 2 step: opp team 

# i think it's easier to temporarily rename the colums to append "_opp"
```

```python
df_grp2.columns = df_grp2.columns.values + '_opp'
```

```python
df_grp2.columns 
```

```python
new_df2 = pd.merge(new_df, df_grp2,
         how='left', 
         left_on=['GAME_ID', 'TEAM_ID_opp'], 
         right_on=['GAME_ID_opp', 'TEAM_ID_opp']      
        )
```

```python
new_df2
```

# Adding new column(s) for stats differentials

```python
new_df2.loc[:,'d_PTS'] = new_df2['PTS'] - new_df2['PTS_opp']
new_df2.loc[:,'d_AST'] = new_df2['AST'] - new_df2['AST_opp']
new_df2.loc[:,'d_REB'] = new_df2['REB'] - new_df2['REB_opp']
new_df2.loc[:,'d_OREB'] = new_df2['OREB'] - new_df2['OREB_opp']
new_df2.loc[:,'d_DREB'] = new_df2['DREB'] - new_df2['DREB_opp']
new_df2.loc[:,'d_FGM'] = new_df2['FGM'] - new_df2['FGM_opp']
new_df2.loc[:,'d_FGA'] = new_df2['FGA'] - new_df2['FGA_opp']
new_df2.loc[:,'d_FG3M'] = new_df2['FG3M'] - new_df2['FG3M_opp']
new_df2.loc[:,'d_FG3A'] = new_df2['FG3A'] - new_df2['FG3A_opp']
new_df2.loc[:,'d_FTM'] = new_df2['FTM'] - new_df2['FTM_opp']
new_df2.loc[:,'d_FTA'] = new_df2['FTA'] - new_df2['FTA_opp']
new_df2.loc[:,'d_TO'] = new_df2['TO'] - new_df2['TO_opp']
new_df2.loc[:,'d_PF'] = new_df2['PF'] - new_df2['PF_opp']
```
