```python
# New column
# make a column that counts home wins in a separate column for easier tracking

#def a custom function
def IsHomeWin(IsHomeTeam, IsWinner):
    if IsHomeTeam == 1 and IsWinner == 1:
        return 1   
    else:
        return 0
    
#apply it with lambda

df['WINS_HOME'] = df.apply(lambda row: IsHomeWin(row.IsHomeTeam, row.WINS), axis = 1)
```

### calculate the features for 2019

```python
df_grp['d_PTS_per_GAME_TEAM1'] = (df_grp['PTS'] - df_grp['PTS_opp'] ) / df_grp['GAMES']

df_grp['c_WIN_pct_TEAM1'] = df_grp['WINS'] / df_grp['GAMES']

df_grp['c_WIN_pct_at_HOME_TEAM1'] = df_grp['WINS_HOME'] / df_grp['IsHomeTeam']

df_grp['c_WIN_pct_at_AWAY_TEAM1'] = (df_grp['WINS'] - df_grp['WINS_HOME'] ) / (df_grp['GAMES'] - df_grp['IsHomeTeam'])

df_grp['c_FGM_per_GAME_TEAM1'] = df_grp['FGM'] / df_grp['GAMES']

df_grp['c_FGA_per_GAME_TEAM1'] = df_grp['FGA'] / df_grp['GAMES']

df_grp['c_PTS_per_GAME_TEAM1'] =  df_grp['PTS'] / df_grp['GAMES']

df_grp['c_PTS_opp_per_GAME_TEAM1'] =  df_grp['PTS_opp'] / df_grp['GAMES']
```

```python
# advanced stats -- possession

df_grp['POS'] = 0.96 * ((df_grp['FGA']) + (df_grp['TO']) + (0.44 * df_grp['FTA']) - (df_grp['OREB']))
df_grp['POS_opp'] = 0.96 * ((df_grp['FGA_opp']) + (df_grp['TO_opp']) + (0.44 * df_grp['FTA_opp']) - (df_grp['OREB_opp']))
```

```python
# advanced stats -- efficiency

df_grp['c_OFF_EFF_TEAM1'] = 100 * (df_grp['PTS'] / df_grp['POS'])
df_grp['c_DEF_EFF_TEAM1'] = 100 * (df_grp['PTS_opp'] / df_grp['POS'])
```

```python
# advanced stats -- last 3/8 games winning pct (use 50%) .. assume playoffs as a "fresh start"
df_grp['r_WIN_pct_L8_TEAM1'] = 0.0
df_grp['r_WIN_pct_L3_TEAM1'] = 0.0
```

### helper functions

```python
def add_features_to_df(df_grps):
    '''
    takes a list of grouped dataframes and adds the features each one.
    returns back a list of the modified dataframes
    '''
    result_dfs = []
    
    for df_grp in df_grps:
        df_grp['d_PTS_per_GAME_TEAM1'] = (df_grp['PTS'] - df_grp['PTS_opp'] ) / df_grp['GAMES']
        df_grp['c_WIN_pct_TEAM1'] = df_grp['WINS'] / df_grp['GAMES']
        df_grp['c_WIN_pct_at_HOME_TEAM1'] = df_grp['WINS_HOME'] / df_grp['IsHomeTeam']
        df_grp['c_WIN_pct_at_AWAY_TEAM1'] = (df_grp['WINS'] - df_grp['WINS_HOME'] ) / (df_grp['GAMES'] - df_grp['IsHomeTeam'])
        df_grp['c_FGM_per_GAME_TEAM1'] = df_grp['FGM'] / df_grp['GAMES']
        df_grp['c_FGA_per_GAME_TEAM1'] = df_grp['FGA'] / df_grp['GAMES']
        df_grp['c_PTS_per_GAME_TEAM1'] =  df_grp['PTS'] / df_grp['GAMES']
        df_grp['c_PTS_opp_per_GAME_TEAM1'] =  df_grp['PTS_opp'] / df_grp['GAMES']
    
        df_grp['POS'] = 0.96 * ((df_grp['FGA']) + (df_grp['TO']) + (0.44 * df_grp['FTA']) - (df_grp['OREB']))
        df_grp['POS_opp'] = 0.96 * ((df_grp['FGA_opp']) + (df_grp['TO_opp']) + (0.44 * df_grp['FTA_opp']) - (df_grp['OREB_opp']))
    
    # advanced stats -- efficiency

        df_grp['c_OFF_EFF_TEAM1'] = 100 * (df_grp['PTS'] / df_grp['POS'])
        df_grp['c_DEF_EFF_TEAM1'] = 100 * (df_grp['PTS_opp'] / df_grp['POS'])
    
    # advanced stats -- last 3/8 games winning pct (use 50%) .. assume playoffs as a "fresh start"
        df_grp['r_WIN_pct_L8_TEAM1'] = 0.0
        df_grp['r_WIN_pct_L3_TEAM1'] = 0.0
        
        result_dfs.append(df_grp)
        
    return result_dfs
```

```python
def filter_features(df_grps):
    
    result_dfs=[]
    
    for df_grp in df_grps:
         
        df_grp_new = df_grp[[

        'd_PTS_per_GAME_TEAM1', 
        'c_WIN_pct_TEAM1',
        'c_WIN_pct_at_HOME_TEAM1', 
        'c_WIN_pct_at_AWAY_TEAM1',
        'c_FGM_per_GAME_TEAM1', 
        'c_FGA_per_GAME_TEAM1',
        'c_PTS_per_GAME_TEAM1',
        'c_PTS_opp_per_GAME_TEAM1', 

        'c_OFF_EFF_TEAM1', 
        'c_DEF_EFF_TEAM1',
        'r_WIN_pct_L3_TEAM1',
        'r_WIN_pct_L8_TEAM1'
        ]].copy()
        result_dfs.append(df_grp_new)
    
    return result_dfs
```

```python
def get_playoff_features(df, team1, team2):
    '''
    takes the dataframe and returns 2 dataframes with the features for each team (using h2h, if available.)
    df is the dataframe containing all regular season matchup outcomes
    team 1 is the higher seeded team
    team 2 is the lower seeded team
    '''
    df_temp1 = df.query("TEAM_ABBR == '" + team1 + "' & OPP_TEAM_ABBR == '" + team2 + "'")
    df_grp1 = df_temp1.groupby('TEAM_ABBR').sum()
    
    df_temp2 = df.query("TEAM_ABBR == '" + team2 + "' & OPP_TEAM_ABBR == '" + team1 + "'")
    df_grp2 = df_temp2.groupby('TEAM_ABBR').sum()
    
    # return df_grp1, df_grp2
    
    temp_list = add_features_to_df([df_grp1, df_grp2])
    
    return filter_features(temp_list)
        
    # df_team1_features, df_team2_features = zip(add_features_to_df([df_grp1, df_grp2]))      
    # return df_team1_features, df_team2_features
```
